```python
# import data
import pandas as pd
df = pd.read_csv('C:/Users/there/notebook/csv files/world happiness dataset/2018.csv')

#rename country or region to country
df.rename(columns = {'Country or region':'Country'}, inplace=True)
df
```

```python
# filter result base on a column
df[df['Overall rank'] > 20]
```

```python
#filter result base on if value are in a column in result
countries = ['Yemen','Nigeria']

df[df['Country'].isin(countries)]
```

```python
#filter result set if a particular string value is in a column
df[df['Country'].str.contains('Niger')]
```

```python
# filter using the filter method
df.filter(items = [0, 1], axis=0)
```

```python
df.filter(like = 'Cou', axis=1)
```
